```python
def swap_matrix(n_qudits, q1, q2, dim=2):
    p1, p2 = sorted([q1, q2])
    SW = zeros([dim**n_qudits, dim**n_qudits])
    SW_ph = zeros([dim, dim])
    for i in range(dim):
        for j in range(dim):
            SW_ph[[i], [j]] = 1
            SW_ij = array(SW_ph)
            SW_ph[[i], [j]] = 0
            SW_ph[[j], [i]] = 1
            SW_ji = array(SW_ph)
            SW_ph[[j], [i]] = 0
            op = reduce(kron, [eye(dim**p1), SW_ij, eye(dim**(p2 - p1 - 1)), SW_ji, eye(dim**(n_qudits - p2 - 1))])
            SW += op    
    return SW
def sp_swap_matrix(n_qudits, q1, q2, dim=2):
    return sp.Matrix(swap_matrix(n_qudits, q1, q2, dim=2))
#sp_swap_matrix(4, 0, 3, dim=2)
```

```python
def ave_haar(negSq, k):
    return (10 + k*negSq)/(20+k)
```

```python
def var_haar(negSq, k):
    return 16*k**2*negSq*(4-negSq)/((51+4*k)**2)
def var_haar_reduced(negSq, k):
    return 16*k**2*negSq*(4-negSq)/((51+4*k)**2)/(4*k/(51+4*k))**2
```

```python
vars_haar=[var_haar_reduced(label, 10) for label in labels_test]
plt.plot(labels_test, vars_haar,label="variance")
plt.plot(labels_test_CRB, 1/np.array(CFIs)+0.01, label="1/CFI")
plt.plot(labels_test_CRB,1/np.array(QFIs), label="1/QFI")
plt.legend()
```

```python
n_copies = 2
n_tot = n_inp*n_copies
w_ls = 1e0
w_var = 1e-4
K=w_ls/w_var
method = "BFGS"
options = {"maxiter": int(1e4)}#, "maxfun": int(1e10)}#

### Don't forget to update theoretical values:
labels_test_haar=[ave_haar(negSq, K) for negSq in labels_test]
disps_test_haar=[var_haar(negSq, K) for negSq in labels_test]
```

```python
ansatz = []
# default case:
ansatz.append(eye(2**n_tot))

# swaps between first pair and second pair
ansatz.append(swap_matrix(n_tot, 0, 2, dim=2)) # this
ansatz.append(swap_matrix(n_tot, 1, 3, dim=2)) # this
ansatz.append(swap_matrix(n_tot, 0, 2, dim=2)@swap_matrix(n_tot, 1, 3, dim=2))

n_pars = len(ansatz)
n_pars
```

```python
dms_train_c = zeros([n_train, 2**n_tot, 2**n_tot], dtype=complex)
for i in range(n_train):
    dms_train_c[i] = reduce(kron, [dms_train[i]]*n_copies)

time_start = time()
Q = zeros((n_train, n_pars))
R = zeros((n_train, n_pars, n_pars))
for i in range(n_train):
    print("Computing Q and R: i=%d" %i, end="\r")
    for j in range(n_pars):
        op_loc = dms_train_c[i]@ansatz[j]
        Q[i][j] = trace(op_loc).real
        for k in range(n_pars):
            R[i][j][k] = trace(op_loc@ansatz[k]).real
time_finish = time() - time_start
print("Computing Q and R: completed in %.2f s" %time_finish, end="\r")
```
